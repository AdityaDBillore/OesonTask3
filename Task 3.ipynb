Here in StatewiseTestingDetails.csv, there are many null values. There are two ways of dealing with null values, viz., either you can drop the null values or you can fill the missing values with mean, median, or mode. But in our dataset, Negative column can be filledd by subtracting positive cases from total samples. The process is shown below

```python
covid_state['Positive'].fillna(0, inplace = True)
```

```python
"Filling negative cases , We counted the values by using values in TotalSamples and positive column"

covid_state['Negative'] = covid_state['TotalSamples'] - covid_state['Positive']
covid_state
```

```python
"Changing the format of date"

covid_india['Date'] = pd.to_datetime(covid_india['Date'])
covid_india['Date'] = covid_india['Date'].dt.strftime('%d-%m-%Y')
covid_india['Date']
```

```python
"Removing unwanted columns from dataset using 'drop'."

covid_india.drop(['Time','ConfirmedIndianNational','ConfirmedForeignNational'],axis = 1,inplace = True)
covid_india
```

```python
"Identifying active cases , We counted the values by using values in confirmed, cured, deaths column"

covid_india['Active'] = covid_india['Confirmed'] - covid_india['Cured'] - covid_india['Deaths']
covid_india
```

```python
"Top 10 states where more sample testing is done"

state_top_10 = covid_state.nlargest(10,['TotalSamples'])

state_top_10 = covid_state.groupby(['State'])['TotalSamples'].max().sort_values(ascending=False).reset_index()
state_top = state_top_10.nlargest(10,['TotalSamples'])
state_top
```

```python
"using pivot function to find cured , deaths , confirmed cases State wise"

statewise = pd.pivot_table(covid_india,values=['Cured','Deaths','Confirmed'],index='State/UnionTerritory',aggfunc='max',margins=True)
statewise
```

```python
"Top 10 states with most Active cases"

covid_top_10 = covid_india.nlargest(10,['Active'])

covid_top_10 = covid_india.groupby(['State/UnionTerritory'])['Active'].max().sort_values(ascending=False).reset_index()
covid_top = covid_top_10.nlargest(10,['Active'])
covid_top
```

```python
"Top 10 states with most deaths cases"

covid_deaths_top_10 = covid_india.nlargest(10,['Deaths'])

covid_deaths_top_10 = covid_india.groupby(['State/UnionTerritory'])['Deaths'].max().sort_values(ascending=False).reset_index()
covid_deaths = covid_deaths_top_10.nlargest(10,['Deaths'])

covid_deaths
```

# Plotting the observations

```python
"""Barplot for Confirmed , Deaths , Cured , Active"""
fig = plt.figure(figsize=(20,10))

Confirmed = covid_india['Confirmed'].sum()
Cured = covid_india['Cured'].sum()
Deaths = covid_india['Deaths'].sum()
Active = covid_india['Active'].sum()

print('Total Confirmed cases =',Confirmed)
print('Total Cured cases =',Cured)
print('Total Active cases =',Active)
print('Total Death cases =',Deaths)

barplot = sns.barplot(x=['Confirmed','Cured','Deaths','active'],y=[Confirmed,Cured,Deaths,Active])
barplot.set_yticklabels(labels=(barplot.get_yticks()*1).astype(int))

plt.show()
```

```python
"Piechart for 'Confirmed','Cured',Deaths & 'Active'"

fig = plt.figure(figsize=(15,10))
covid_values = [covid_india['Confirmed'].sum(),covid_india['Cured'].sum(),covid_india['Deaths'].sum(),covid_india['Active'].sum()]
covid_keys = [Confirmed,Cured,Deaths,Active]


plt.pie(covid_keys,labels = covid_keys, explode = (0.02,0.02,0.1,0.02), autopct = '%.0f%%')
plt.legend(['Confirmed','Cured','Deaths','Active'])
```

```python
"Pie Chart Of 10 Top states with Active Cases"

fig = plt.figure(figsize=(13,10))
covid_top.groupby(["State/UnionTerritory"]).sum()["Active"].plot(kind='pie',rot=90,explode=(0.05,0.02,0.03,0.04,0.04,0.05,0.1,0.04,0.09,0.04),autopct='%1.0f%%')
plt.title('Top 10 states with most Active cases',size=20)
plt.show()
```

```python
"Bar Plot Of Top 10 Active Cases"

fig = plt.figure(figsize=(20,10))
sns.barplot(data = covid_top.iloc[:10],y='Active',x='State/UnionTerritory')
plt.title('Top 10 states with Most Active cases', size=20)
plt.show()
```

```python
" Top 5 Most affected states"

fig = plt.figure(figsize=(15,10))
plot = sns.lineplot(data = covid_india[covid_india['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Tamil Nadu','Delhi','Uttar Pradesh'])],x='Date',y='Active',hue = 'State/UnionTerritory',size='State/UnionTerritory')
plt.title('5 most affected states',size=20)
plt.show()
```

```python
"Correlation Heatmap"

fig = plt.figure(figsize=(15,10))
sns.heatmap(covid_india.corr(),cmap="Reds",annot=True)
plt.title('Correlation Heatmap',size=20)
plt.show()
```

```python
"Correlation Heatmap"

fig = plt.figure(figsize=(15,10))
sns.heatmap(covid_state.corr(),cmap="Reds",annot=True)
plt.title('Correlation Heatmap',size=20)
plt.show()
```

```python
"Fatality ratio of contaminated states"

covid_india['Fatality_ratio']= covid_india['Deaths']/covid_india['Confirmed']
fig,ax = plt.subplots(figsize=(15,7))
sns.pointplot(data = covid_india,x='State/UnionTerritory',y='Fatality_ratio',ax=ax,color='Red')
plt.xticks(rotation=90)
plt.title('Fatality ratio of contaminated states',size=20)
plt.show()
```
